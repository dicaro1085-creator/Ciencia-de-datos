```python
import numpy as np

# Generar un array NumPy con datos financieros simulados (Precio acciones por dias)

# Número de días a simular
num_days = 100

# Precio inicial de la acción
initial_price = 50.0

# Generar cambios porcentuales diarios aleatorios (pequeñas fluctuaciones)
daily_returns = np.random.normal(0.001, 0.01, num_days) # Media de 0.1% de retorno, desviación estándar de 1%

# Calcular los precios acumulados
simulated_prices = initial_price * np.exp(np.cumsum(daily_returns))

print("Precios de acciones simulados (primeros 10 días):")
print(simulated_prices[:10])
print(f"\nTotal de {num_days} precios simulados.")

```

```python
import numpy as np

# Tomar los primeros 25 elementos para formar la matriz 5x5
prices_for_matrix = simulated_prices[:25]

# Organizar los datos en una matriz de 5x5
stock_prices_matrix = prices_for_matrix.reshape(5, 5)

print("Matriz de precios de acciones (5x5):")
print(stock_prices_matrix)
```

```python
import numpy as np

print("Estadísticas por acción (fila):")
for i, row in enumerate(stock_prices_matrix):
    avg_price = np.mean(row)
    max_price = np.max(row)
    min_price = np.min(row)
    print(f"  Acción {i+1}: Promedio = {avg_price:.2f}, Máximo = {max_price:.2f}, Mínimo = {min_price:.2f}")
```

```python
import numpy as np

print("Variación porcentual diaria por acción (fila):")
for i, row in enumerate(stock_prices_matrix):
    # Calcula la variación porcentual diaria
    # (Precio_actual - Precio_anterior) / Precio_anterior * 100
    daily_percentage_change = (np.diff(row) / row[:-1]) * 100
    # Aquí lo imprimimos directamente para mostrar todos los valores.
    print(f"  Acción {i+1}: {daily_percentage_change}")
```

```python
import numpy as np

print("Aplicando funciones matemáticas a stock_prices_matrix:")

# Aplicar logaritmo natural (np.log) a los precios de las acciones
# Es importante asegurarse de que no haya valores <= 0 antes de aplicar log
log_stock_prices = np.log(stock_prices_matrix)
print("\nMatriz de precios de acciones (logaritmo natural):")
print(log_stock_prices)

# Aplicar función exponencial (np.exp) a los precios de las acciones
# (Esto "deshace" un logaritmo, o simplemente muestra el efecto de exp)
exp_stock_prices = np.exp(stock_prices_matrix)
print("\nMatriz de precios de acciones (exponencial):")
print(exp_stock_prices)

# Normalización Min-Max para cada acción (fila)
normalized_stock_prices = np.zeros_like(stock_prices_matrix, dtype=float)
for i, row in enumerate(stock_prices_matrix):
    min_val = np.min(row)
    max_val = np.max(row)
    # Evitar división por cero si max_val == min_val
    if (max_val - min_val) != 0:
        normalized_stock_prices[i, :] = (row - min_val) / (max_val - min_val)
    else:
        normalized_stock_prices[i, :] = 0.0 # O 0.5 si todos los valores son iguales y se desea el centro

print("\nMatriz de precios de acciones (Normalización Min-Max por fila):")
print(normalized_stock_prices)
```

```python
import numpy as np

print("Extrayendo información específica con indexación avanzada de stock_prices_matrix:")

# Rendimiento de la primera acción (fila 0) en el tercer día (columna 2)
# La indexación en Python/NumPy comienza en 0
accion_1_dia_3 = stock_prices_matrix[0, 2]
print(f"\nPrecio de la Acción 1 en el Día 3: {accion_1_dia_3:.2f}")

# Rendimiento de la segunda acción (fila 1) en todos los días (todas las columnas)
accion_2_todos_dias = stock_prices_matrix[1, :]
print("\nPrecios de la Acción 2 en todos los días:")
print(accion_2_todos_dias)

# Rendimiento de todas las acciones en el cuarto día (columna 3)
todas_acciones_dia_4 = stock_prices_matrix[:, 3]
print("\nPrecios de todas las Acciones en el Día 4:")
print(todas_acciones_dia_4)

# Rendimiento de las primeras tres acciones (filas 0, 1, 2) durante los primeros dos días (columnas 0, 1)
sub_matriz_rendimiento = stock_prices_matrix[0:3, 0:2]
print("\nPrecios de las primeras 3 Acciones durante los primeros 2 Días:")
print(sub_matriz_rendimiento)
```

```python
import numpy as np

print("Demostrando Broadcasting con stock_prices_matrix:")

# --- Ejemplo 1: Sumar un valor constante a todos los precios ---
# Esto añade 5 a cada elemento de la matriz
constant_addition = stock_prices_matrix + 5
print("\nMatriz con un valor constante (5) sumado a cada elemento:")
print(constant_addition)

# --- Ejemplo 2: Multiplicar cada fila (acción) por un factor diferente ---
# Creamos un array unidimensional de factores, uno por cada acción (fila).
# stock_prices_matrix tiene 5 filas, así que necesitamos 5 factores.
factors = np.array([1.01, 0.98, 1.05, 0.99, 1.02]) # Factores arbitrarios

# NumPy automáticamente 'estira' el array de factores para que coincida
# con las filas de stock_prices_matrix y realiza la multiplicación.
# El array 'factors' se transmite a lo largo de las columnas.
scaled_prices = stock_prices_matrix * factors[:, np.newaxis]
print("\nMatriz donde cada acción (fila) ha sido multiplicada por un factor diferente:")
print(scaled_prices)
```

Comparacion entre Numpy y Python

Eficiencia codigo

Numpy: Conciso y legible, las operaciones se expresan a menudo en una sola línea utilizando funciones vectorizadas.

Python: Requiere el uso explícito de bucles (for) para la mayoría de las operaciones, especialmente en datos multidimensionales (listas anidadas).

Rendimiento Computacional

Numpy: Altamente optimizado, lo que permite que los cálculos se realicen mediante rutinas de bajo nivel precompiladas. Las operaciones de NumPy están vectorizadas, lo que significa que operan en arrays completos a la vez, eliminando la sobrecarga de los bucles de Python. Los arrays de NumPy almacenan los datos en bloques contiguos de memoria, lo que mejora el rendimiento de la caché y el acceso a la memoria. Muchas operaciones subyacentes de NumPy pueden aprovechar procesadores multi-núcleo o hardware especializado (como GPUs) para la computación paralela, aumentando aún más la velocidad.

Python: Las operaciones sobre listas de Python son generalmente mucho más lentas, especialmente para conjuntos de datos grandes, debido a que Python es un lenguaje interpretado y sus bucles se ejecutan secuencialmente. Almacenan referencias a objetos que pueden estar dispersos en la memoria, lo que es menos eficiente.

Conclusion

Para cálculos numéricos, especialmente con grandes conjuntos de datos y arrays/matrices multidimensionales, NumPy ofrece un rendimiento computacional vastamente superior y un código significativamente más eficiente, legible y conciso en comparación con la implementación de estas operaciones desde cero utilizando listas y bucles de Python puro. Es el estándar de facto para la computación numérica en Python por una razón.
