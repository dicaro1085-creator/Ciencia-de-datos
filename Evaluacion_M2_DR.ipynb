```python
#Proyecto Sistema de Gestión de Contactos
class Contacto:
    def __init__(self, nombre, telefono, correo, direccion):
        self.nombre = nombre
        self.telefono = telefono
        self.correo = correo
        self.direccion = direccion

    def to_dict(self):
        return {
            "nombre": self.nombre,
            "telefono": self.telefono,
            "correo": self.correo,
            "direccion": self.direccion
        }
```

Crear Clase Contacto: representa un contacto con nombre, telefono, correo, direccion
Usamos guion bajo (__) para indicar que los atributos son "protegidos"
lo cual indica que tiene encapsulación, ayuda a proteger los datos importantes de cambios no deseados y a mantener el código organizado.

def to_dict(self) Define un método dentro de la clase "Contacto" y convierte el objeto en un diccionario esto facilita la visualización y manipulación de los datos.


```python
#Gestor de contactos
class GestorContactos:
    def __init__(self):
        self.contactos = []

    def agregar_contacto(self, nombre, telefono, correo, direccion):
        contacto = Contacto(nombre, telefono, correo, direccion)
        self.contactos.append(contacto)

    def listar_contactos(self):
        return [c.to_dict() for c in self.contactos]

    def buscar_contacto(self, dato):
        resultados = []
        for contacto in self.contactos:
            if dato.lower() in contacto.nombre.lower() or dato in contacto.telefono:
                resultados.append(contacto.to_dict())
        return resultados

    def editar_contacto(self, telefono, nombre, correo, direccion):
        for contacto in self.contactos:
            if contacto.telefono == telefono:
                contacto.nombre = nombre
                contacto.correo = correo
                contacto.direccion = direccion
                return True
        return False

    def eliminar_contacto(self, telefono):
        for contacto in self.contactos:
            if contacto.telefono == telefono:
                self.contactos.remove(contacto)
                return True
        return False
```

se crea una clase llamada "GestorContactos" que administra todos los contactos.

def__init__(self) prepara los datos
[] indica la lista vacia

def agregar_contacto: define las funciones del gestor de contactos
agregar_contacto: crea un nuevo contacto usando la clase contacto y lo guarda dentro de la lista self.contacto

listar_contactos:muestra todos los contactos guardados

buscar_contacto:busca contactos por nombre o numero de telefono

editar_contacto:modifica la información de un contacto existente

eliminar_contacto:elimina un contacto del sistema

```python
def menu():
  print("1. Agregar Contacto")
  print("2. Listar Contactos")
  print("3. Buscar Contacto")
  print("4. Editar Contacto")
  print("5. Eliminar Contacto")
  print("6. Salir")

gestor = GestorContactos()

while True:
    menu()
    opcion = input("Selecciona una opción: ")
    if opcion == "1":
        nombre = input("Nombre: ")
        telefono = input("Teléfono: ")
        correo = input("Correo: ")
        direccion = input("Dirección: ")
        gestor.agregar_contacto(nombre, telefono, correo, direccion)
        print("Contacto agregado.")
    elif opcion == "2":
        contactos = gestor.listar_contactos()
        if not contactos:
            print("Contacto no encontrado.")
        else:
            print("Lista de Contactos:")
            for contacto in contactos:
                print(contacto)
    if opcion == "3":
        dato = input("Nombre o Teléfono: ")
        resultados = gestor.buscar_contacto(dato)
        if not resultados:
            print("Contacto no encontrado.")
        else:
            print("Resultados de la búsqueda:")
            for resultado in resultados:
                print(resultado)
    if opcion == "4":
        telefono = input("Teléfono del contacto a editar: ")
        nombre = input("Nuevo nombre: ")
        correo = input("Nuevo correo: ")
        direccion = input("Nueva dirección: ")
        if gestor.editar_contacto(telefono, nombre, correo, direccion):
            print("Contacto editado.")
        else:
            print("Contacto no encontrado.")
    if opcion == "5":
        telefono = input("Teléfono del contacto a eliminar: ")
        if gestor.eliminar_contacto(telefono):
            print("Contacto eliminado.")
        else:
            print("Contacto no encontrado.")

    elif opcion == "6":
        print("Saliendo del programa.")
        break
    else:
        print("Opción no válida. Ingresé una opción del Menú. \n")

```
